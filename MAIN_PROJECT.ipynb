**DEEP LEARNING MODEL TO PREDICT WHETHER THE CANCER IS MALIGNANT OR BENIGN**

IMPORTING THE DATASET

```python
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()
data
```

SPLITTING THE FEATURE AND TARGET VALUES FROM THE DATASET

```python
features = data.data
targets = data.target
```

FINDING THE SHAPES OF THE FEATURE AND TARGET VALUES

```python
print(features.shape)
print(targets.shape)
```

SPLITTING OF TRAINING AND TESTING DATA

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(features,targets,test_size=0.2)
```

FINDING THE SHAPES OF THE TRAINING AND TESTING DATA

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

CREATING A NEURAL NETWORK

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model=Sequential()
model.add(Dense(units=30,activation="relu",input_dim=30))
model.add(Dense(units=60,activation="relu"))
model.add(Dense(units=1,activation="sigmoid"))
model.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
model.summary()
```

TRAINING THE DATA

```python
model.fit(x_train,y_train,epochs=500)
```

EVALUATING THE TEST DATA

```python
model.evaluate(x_test,y_test)
```

**ACCURACY OF OUR MODEL IS 98%**

PREDICTING WHETHER THE CANCER IS MALIGNANT OR BENIGN

```python
predictions = model.predict(x_test)
label = []
for prediction in predictions:
  if(prediction<0.5):
    label.append("BENIGN")
  else:
    label.append("MALIGNANT")

label
```
